```python
dtf = pd.DataFrame() #Создаем пустой дата фрейм, куда будем сводить отчеты
```

```python
def trend_data(text_file_name, dtf):
  df = pd.read_csv(text_file_name, sep = ': ')
  df = df.set_index('Возраст').T
  dtf = pd.concat([
      dtf,
      df
      ])
  return dtf
```

```python
text_file_name = input("Вставить путь файла: ")
dtf = trend_data(text_file_name, dtf) #Создаем БД тенденций
dtf.head() #Проверка
```

```python
dtf.to_excel('Тренды.xlsx') #Выводим таблицу Exel
```
