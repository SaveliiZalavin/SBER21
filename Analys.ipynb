Создаем таблицу на основе полученных отчетов

```python
dtf = pd.DataFrame() #Создаем пустой дата фрейм, куда будем сводить отчеты
```

```python
def trend_data(text_file_name, dtf):
  df = pd.read_csv(text_file_name, sep = ': ')
  df = df.set_index('Возраст').T
  dtf = pd.concat([
      dtf,
      df
      ])
  return dtf
```

```python
text_file_name = input("Вставить путь файла: ")
dtf = trend_data(text_file_name, dtf) #Создаем БД тенденций
dtf.head() #Проверка
```

```python
dtf.to_excel('Тренды.xlsx') #Выводим таблицу Exel
```

```python
def analyse(file_way):
  df = pd.read_excel(file_way)
  #Заменяем пустые поля на нули
  df = df.fillna(0)
  df.index = df['2. Сколько Вам лет?']
  #Исключаем до 18 лет и 35+ лет
  df = df.loc[lambda df1: df1.index != 'до 18']
  df = df.loc[lambda df1: df1.index != '35+']
  #Подсчет количества отметивших вариант о каких страховых знают по возрасту
  df1 = df[['Знают про Альфу',
            'Знают про Тинькофф',
            'Знают про СБЕР',
            'Знают про Газпром',
            'Знают про МТС']].copy()
  df1.index = df['2. Сколько Вам лет?']
  df1 = df1.replace(['Да'], 1)
  df1 = df1.replace(['Нет'], 0)
  df1 = df1.groupby('2. Сколько Вам лет?').sum()
  df1.plot.bar()
  plt.title(f'Про какие страховые знают')
  #Подсчет количества откуда узнают
  df2 = df[['Источник знакомые',
            'Источник турфирмы',
            'Источник интернет и отзывы',
            'Источник наружная реклама',
            'Источник видео',
            'Источник контекстная']].copy()
  df2.index = df['2. Сколько Вам лет?']
  df2 = df2.replace(['Да'], 1)
  df2 = df2.replace(['Нет'], 0)
  df2 = df2.groupby('2. Сколько Вам лет?').sum()
  df2.plot.bar()
  plt.title(f'Откуда узнают')
  #Подсчет количества что важно при выборе страховой компании
  df3 = df[['Важно известная надежная компания',
            'Важно соотношение цена/качество',
            'Важно удобство при оформлении',
            'Важно креатив',
            'Важно бонусы и рефералки',
            'Важно скорость оформления',
            'Важно положительные отзывы']].copy()
  df3.index = df['2. Сколько Вам лет?']
  df3 = df3.replace(['Да'], 1)
  df3 = df3.replace(['Нет'], 0)
  df3 = df3.groupby('2. Сколько Вам лет?').sum()
  df3.plot.bar()
  plt.title(f'Что важно при выборе страховой компании')
  #Подсчет количества отметивших вариант что может помешать
  df4 = df[['Отталкивает незнание',
            'Отталкивает не считают нужным',
            'Отталкивает недостаток информации',
            'отталкивает неудобство оформления',
            'Отталкивает неструктурированность',
            'Отталкивает устаревший интерфейс',
            'Отталкивает отсутствие нужных опций',
            'Отталкивает отсутствие поддержки',
            'Отталкивает плохие отзывы']].copy()
  df4.index = df['2. Сколько Вам лет?']
  df4 = df4.replace(['Да'], 1)
  df4 = df4.replace(['Нет'], 0)
  df4 = df4.groupby('2. Сколько Вам лет?').sum()
  df4.plot.bar()
  plt.title(f'Барьеры для потребителя')
  #Подсчет количества отметивших важные опции
  df5 = df[['Опция переводчик',
            'Опция сохранение документов офлайн',
            'Опция синхронизации с госуслугами',
            'Опция кастомизации тарифа']].copy()
  df5.index = df['2. Сколько Вам лет?']
  df5 = df5.replace(['Да'], 1)
  df5 = df5.replace(['Нет'], 0)
  df5 = df5.groupby('2. Сколько Вам лет?').sum()
  df5.plot.bar()
  plt.title(f'Важные опции')
  #Подсчет количества отметивших вариант что важно в самой страховке
  df6 = df[['Значимо широкое покрытие',
            'Значимо 24/7 поддержка',
            'Значимо широкая география',
            'Значимо высокая страховая сумма',
            'Значимо помощь в коммуникации',
            'Значимо защита багажа',
            'Значимо задержка/отмена рейса']].copy()
  df6.index = df['2. Сколько Вам лет?']
  df6 = df6.replace(['Да'], 1)
  df6 = df6.replace(['Нет'], 0)
  df6 = df6.groupby('2. Сколько Вам лет?').sum()
  df6.plot.bar()
  plt.title(f'Важно в самой страховке')
  #Подсчет количества отметивших вариант что хотель бы включить в страховку
  df7 = df[['Включить отмена поездки по УП',
            'Включить защита багажа',
            'Включить задержка/отмена рейса',
            'Включить защита жилья на время отсутствия',
            'Помощь на дороге',
            'Включить утрата документов',
            'Включить несчастный случай']].copy()
  df7.index = df['2. Сколько Вам лет?']
  df7 = df7.replace(['Да'], 1)
  df7 = df7.replace(['Нет'], 0)
  df7 = df7.groupby('2. Сколько Вам лет?').sum()
  df7.plot.bar()
  plt.title(f'Включить в страховку')
  #Подсчет количества отметивших вариант ассоциации с идеальной страховой
  df8 = df[['Ассоциация надежная',
             'Ассоциация выгодное',
             'Ассоциация удобное',
             'Ассоциация качественное',
             'Ассоциация индивидуальное',
             'Ассоциация эффективное',
             'Ассоциация комплексное',
             'Ассоциация универсальное',
             'Ассоциация доступное']].copy()
  df8.index = df['2. Сколько Вам лет?']
  df8 = df8.replace(['Да'], 1)
  df8 = df8.replace(['Нет'], 0)
  df8 = df8.groupby('2. Сколько Вам лет?').sum()
  df8.plot.bar()
  plt.title(f'Ассоциация с идеальной страховой')
```
